```python
import pandas as pd
import qnorm
import matplotlib.pyplot as plt
```

```python
stages = ["Eggs", "2-4cells", "2hpf", "1Kcells", "Dome", "Shield", "8hpf", "Bud", "12hpf", "16hpf", "20hpf", "26hpf", "28hpf", "48hpf", "72hpf", "120hpf", "7dpf"]

table = []

df_1 = pd.read_table(f"../data_new/GRCz11-TPM.tsv.gz", index_col=0)


for stage in stages:
    reps = [col for col in df_1.columns if col.startswith(stage)]
    
    df = pd.DataFrame(df_1[reps].mean(axis=1))
    df.columns = [stage]
    table.append(df)

df_1 = pd.concat(table, axis=1)
df_1 = qnorm.quantile_normalize(df_1)
df_1
```

```python
stages = ["1-cell", "2-cell", "128-cell", "1k-cell", "Dome", "50pc-epiboly", "Shield", "75pc-epiboly", "1-4-somites", "14-19-somites", "20-25-somites", "Prim-5", "Prim-15", "Prim-25", "Long-pec", "Protruding-mouth", "Day-4", "Day-5"]

table = []

df_2 = pd.read_table(f"../data_new/GRCz11-TPM_singleembryo.tsv.gz", index_col=0)

for stage in stages:
    reps = [col for col in df_2.columns if col.startswith(stage)]
    
    df = pd.DataFrame(df_2[reps].mean(axis=1))
    df.columns = [stage]
    table.append(df)

df_2 = pd.concat(table, axis=1)
df_2 = qnorm.quantile_normalize(df_2)
df_2
```

```python
cur = gen_orthodb("../data_new/orthogroups.tsv")

orthogroups = []

for gene in df_1.index:
    res = cur.execute(
    f"""
    SELECT genes.orthogroup from 
    ORTHOGROUPS INNER JOIN genes ON ORTHOGROUPS.orthogroup=genes.orthogroup
    WHERE genes.assembly='GRCz11' AND 
    (genes.gene_name ='{gene}' OR genes.gene_id='{gene}')
    """).fetchall()

    assert len(res) <= 1
    if len(res) == 1:
        orthogroups.append((res[0][0], gene))
    else:
        orthogroups.append((None, gene))
        
# df.index = pd.MultiIndex.from_tuples(orthogroups)
# df.index.set_names(["orthogroup", "gene"], inplace=True)
```

```python
ortho_df = pd.read_table("../data_new/orthogroups.tsv", index_col=0)
ortho_df

keep = []

for orthogroup, gene in orthogroups:
    if orthogroup == None:
        continue
    lookup = ortho_df.loc[orthogroup, "BraLan3.pep"]
    lookup = str(lookup)
    if lookup != "nan" and ", " in lookup:
        continue

    lookup = ortho_df.loc[orthogroup, "GRCz11.pep"]
    lookup = str(lookup)
    if ", " in lookup:
        continue
    
    keep.append(gene)
    
```

```python
df_1 = qnorm.quantile_normalize(df_1.loc[keep])
df_2 = qnorm.quantile_normalize(df_2.loc[keep])
df_3 = qnorm.quantile_normalize(df_3.loc[keep])
```

```python
fig, axes = plt.subplots(3, 3, figsize=(5, 5))

for i, df_X in enumerate([df_1, df_2, df_3]):
    for j, df_Y in enumerate([df_1, df_2, df_3]):
        if j > i:
            axes[i, j].axis('off')
            continue
        corrs = pairwise_jensen_shannon(df_X, df_Y)
        axes[i, j].imshow(corrs, aspect="auto", cmap="RdBu_r") #, vmin=0, vmax=1)
```
